```python
import pandas as pd
import folium
import geopandas as gpd
```

```python
data = pd.read_csv('/mnt/large/data/PseudoPFLOW/activity/13/person_13101_labor.csv', names=['id', 'age', 'gender', 'occupation', 'starttime', 'duration', 'purpose', 'lon', 'lat', 'code'])
data
```

```python
map_center = [data['lat'].mean(), data['lon'].mean()]
m = folium.Map(location=map_center, zoom_start=14)
m
```

```python
for idx, row in data.iterrows():
    if row['purpose'] not in [1,2,3,400, 500]:
        folium.Marker([row['lat'], row['lon']], popup=str(row['lat'])+','+str(row['lon'])+','+str(row['purpose'])).add_to(m)
    if idx > 500:
        break
m
```
